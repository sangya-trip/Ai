```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("TopTracksPlaylistsUSA.csv")

freq_table = df['Artist'].value_counts()

print("Frequency Table:")
print(freq_table)

freq_df = freq_table.reset_index()
freq_df.columns = ['Artist', 'count']

plt.figure(figsize=(10,6))
sns.barplot(data=freq_df.head(10), x='count', y='Artist', palette='viridis')

plt.title('Top 10 Artists Frequency')
plt.xlabel('Count')
plt.ylabel('Artist')
plt.show()
```

```python
This Bar chart shows that Sabrina Carpenter was the top artist and Kendrick Lamar was the least artist listened to as per the data from Top Tracks Playlist USA 2024.
```

```python
import pandas as pd

df = pd.read_csv('TopTracksPlaylistsUSA.csv')

relative_freq = df['AI Genre'].value_counts(normalize=True)

print(relative_freq)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

relative_freq_df = relative_freq.reset_index()
relative_freq_df.columns = ['AI Genre', 'relative_frequency']

sns.barplot(data=relative_freq_df, x='AI Genre', y='relative_frequency')
plt.xticks(rotation=45)
plt.title('Relative Frequency of Genres')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

artist_counts = df['Artist'].value_counts().reset_index()
artist_counts.columns = ['Artist', 'count']


top_artists = artist_counts.head(10)

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=top_artists, x='Artist', y='count', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Most Frequently Appeared Artists')

ax.bar_label(ax.containers[0])

plt.tight_layout()
plt.show()
```
