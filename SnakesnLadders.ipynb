# Snakes and Ladders
# Given that a player rolled an even number on their first turn, what is the probability that they land on the first ladder?


This is a theoretical study. We are only observing the data that is produced by the simulator. 

The sample size I will be observing is all the even numbers rolled on the dice.
Simulating one first turn roll of the die, but only recording the outcome if the roll is even (2, 4, or 6).Then, checking if that roll results in landing on the first ladder.
We are measuring whether the player lands on the first ladder or not.

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters
first_ladder_start = 4
num_simulations = 100

def simulate_once(trial_num):
    while True:
        roll = random.randint(1, 6)
        if roll % 2 == 0:
            break
    landed = roll == first_ladder_start
    result_str = "Landed on ladder" if landed else "No ladder"
    print(f"Trial {trial_num}: Roll={roll} → {result_str}")
    return 1 if landed else 0

def run_simulation(n):
    results = []
    for i in range(1, n + 1):
        result = simulate_once(i)
        results.append(result)
    return results

# Run simulation
results = run_simulation(num_simulations)

# Calculate empirical probability
empirical_prob = sum(results) / len(results)

# Theoretical probability calculation
theoretical_prob = 1 / 3

print("\n--- Summary ---")
print(f"Theoretical Probability of landing on ladder given even roll: {theoretical_prob:.4f} ({theoretical_prob*100:.2f}%)")
print(f"Empirical Probability from simulation: {empirical_prob:.4f} ({empirical_prob*100:.2f}%)")

# Visualization
sns.countplot(x=results)
plt.xticks([0, 1], ['No Ladder', 'Landed on Ladder'])
plt.title(f'Results of {num_simulations} Simulations\nLanding on First Ladder Given Even Roll')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

If you know the player rolled an even number on their first turn, then the possible rolls are 2, 4, or 6. To land on the first ladder, which starts on square 4, the player has to roll exactly a 4. Since there are three even numbers they could have rolled, and only one of them gets them onto the ladder, the chance is 1 out of 3 — about 33%. So basically, knowing the roll was even makes it more likely to land on that ladder compared to just rolling the die without any condition.
