# Are books with a rating 4.8 or more likely to belong to a specific genre?

I will use the data set my teacher has provided on Schoology on books to get a scatterplot or a bar chart. 

I will then use the data from my scatterplot or a bar chart to get my analysis done. 

This is an observational study. 

There might be a bias while doing the selection of the books (i.e picking only the ones popular and well known and leaving the others behind. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# Load the dataset
books = pd.read_csv('books.csv')  # Adjust path if needed

# Clean data: ensure 'User Rating' is numeric, drop rows with missing 'User Rating' or 'Genre'
books['User Rating'] = pd.to_numeric(books['User Rating'], errors='coerce')
books = books.dropna(subset=['User Rating', 'Genre'])

print(f"Total rows after cleaning: {len(books)}")  # Check data size (should be 50+)

# Create a new column for high rating (User Rating >= 4.8)
books['High Rated'] = books['User Rating'] >= 4.8

# --- Analysis Technique 1: Frequency Table ---
contingency_table = pd.crosstab(books['Genre'], books['High Rated'])
print("\nFrequency Table (Genre vs High Rated):")
print(contingency_table)

# --- Analysis Technique 2: Relative Frequency Table ---
rel_freq_table = contingency_table.div(contingency_table.sum(axis=1), axis=0)
print("\nRelative Frequency Table (Proportion of High Rated by Genre):")
print(rel_freq_table)

# --- Analysis Technique 3: Pie Chart ---
high_rated_counts = books[books['High Rated']]['Genre'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(high_rated_counts, labels=high_rated_counts.index, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title("Genre Distribution of Books with Rating ≥ 4.8")
plt.axis('equal')
plt.show()

# --- Analysis Technique 4 & 5: Mean and Median of User Ratings ---
mean_rating = books['User Rating'].mean()
median_rating = books['User Rating'].median()
print(f"\nMean User Rating: {mean_rating:.3f}")
print(f"Median User Rating: {median_rating:.3f}")

# --- Analysis Technique 6: Box Plot ---
plt.figure(figsize=(12,6))
sns.boxplot(x='Genre', y='User Rating', data=books)
plt.xticks(rotation=45, ha='right')
plt.title('User Rating Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.tight_layout()
plt.show()

# --- Analysis Technique 7: Z-Scores ---
books['Z-Score'] = zscore(books['User Rating'])
print("\nSample Z-Scores for User Ratings:")
print(books[['User Rating', 'Z-Score']].head())

# Optional: Bar plot of proportion of high-rated books by genre for clearer visualization
prop_high_rated = rel_freq_table[True].sort_values(ascending=False)
plt.figure(figsize=(10,6))
prop_high_rated.plot(kind='bar', color='skyblue')
plt.ylabel('Proportion of Books with Rating ≥ 4.8')
plt.title('Proportion of High-Rated Books by Genre')
plt.show()
```
