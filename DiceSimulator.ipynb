# Dice Simulator

```python
import random
import matplotlib.pyplot as plt
from collections import Counter

# Parameters
num_rolls = 100
sides = 6

# Step 1: Simulate the dice rolls and print them
rolls = []
print(f"ðŸŽ² Rolling a {sides}-sided die {num_rolls} times:\n")

for i in range(num_rolls):
    roll = random.randint(1, sides)
    rolls.append(roll)
    print(f"Roll {i+1}: {roll}")

# Step 2: Count occurrences of each face
roll_counts = Counter(rolls)

# Ensure all faces from 1 to 6 are present
for face in range(1, sides + 1):
    if face not in roll_counts:
        roll_counts[face] = 0

# Sort by face
roll_counts = dict(sorted(roll_counts.items()))

# Step 3: Plot the results with different bar colors
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']  # one color per face

plt.figure(figsize=(8, 5))
bars = plt.bar(roll_counts.keys(), roll_counts.values(), color=colors, edgecolor='black')

plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title(f'Distribution of {num_rolls} Rolls of a {sides}-Sided Die')
plt.xticks(range(1, sides + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from collections import Counter
import matplotlib.cm as cm
import numpy as np

# ====== CONFIGURATION ======
N = 20            # Number of sides on the die (change this to any integer â‰¥2)
num_rolls = 100   # Number of rolls
# ===========================

# Step 1: Roll the die N times
rolls = []
print(f"ðŸŽ² Rolling a {N}-sided die {num_rolls} times:\n")

for i in range(num_rolls):
    roll = random.randint(1, N)
    rolls.append(roll)
    print(f"Roll {i+1}: {roll}")

# Step 2: Count frequencies
roll_counts = Counter(rolls)

# Ensure all die faces are represented (even if frequency is 0)
for face in range(1, N + 1):
    if face not in roll_counts:
        roll_counts[face] = 0

# Sort by die face
roll_counts = dict(sorted(roll_counts.items()))

# Step 3: Generate distinct colors using a colormap
color_map = cm.get_cmap('tab20', N)  # up to 20 distinct colors
colors = [color_map(i) for i in range(N)]

# Step 4: Plot the distribution
plt.figure(figsize=(max(8, N // 2), 5))  # Scale width for larger N
bars = plt.bar(roll_counts.keys(), roll_counts.values(), color=colors, edgecolor='black')

plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title(f'Distribution of {num_rolls} Rolls of a {N}-Sided Die')
plt.xticks(range(1, N + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
