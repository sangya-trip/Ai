```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("pokemondataset - all.csv")

gen1 = df[df['generation'] == 1]

mean_speed = gen1['speed'].mean()
median_speed = gen1['speed'].median()

print(f"Mean Speed: {mean_speed:.2f}")
print(f"Median Speed: {median_speed:.2f}")

plt.figure(figsize=(10, 6))
sns.histplot(data=gen1, x='speed', bins=10, kde=True, color='skyblue', edgecolor='black')

plt.axvline(mean_speed, color='green', linestyle='--', label=f'Mean = {mean_speed:.1f}')
plt.axvline(median_speed, color='orange', linestyle='--', label=f'Median = {median_speed:.1f}')

plt.title("Histogram of Speed (Gen 1 Pokmon)")
plt.xlabel("Speed")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("pokemondataset - all.csv")

# Filter to Generation 1 only
gen1 = df[df['generation'] == 1]

# Mean and median of Gen 1 speed
mean_speed = gen1['speed'].mean()
median_speed = gen1['speed'].median()

print(f"Mean Speed: {mean_speed:.2f}")
print(f"Median Speed: {median_speed:.2f}")

# Get speed of Oddish, Gloom, Vileplume
oddish_speed = df[df['name'].str.lower() == 'oddish']['speed'].values[0]
gloom_speed = df[df['name'].str.lower() == 'gloom']['speed'].values[0]
vileplume_speed = df[df['name'].str.lower() == 'vileplume']['speed'].values[0]

print(f"\nOddish Speed: {oddish_speed}")
print(f"Gloom Speed: {gloom_speed}")
print(f"Vileplume Speed: {vileplume_speed}")

# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=gen1, x='speed', bins=10, kde=True, color='skyblue', edgecolor='black')

# Add lines
plt.axvline(mean_speed, color='green', linestyle='--', label=f'Mean = {mean_speed:.1f}')
plt.axvline(median_speed, color='orange', linestyle='--', label=f'Median = {median_speed:.1f}')
plt.axvline(oddish_speed, color='red', linestyle='-', linewidth=2, label=f'Oddish = {oddish_speed}')
plt.axvline(gloom_speed, color='purple', linestyle='--', label=f'Gloom = {gloom_speed}')
plt.axvline(vileplume_speed, color='blue', linestyle='--', label=f'Vileplume = {vileplume_speed}')

# Labeling
plt.title("Histogram of Speed (Gen 1 Pokmon) with Oddish and Evolutions")
plt.xlabel("Speed")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The Oddish's speed is very low compared to other first gen Pokemons. The speed of Oddish is 30 while the average speed of the other first gen Pokemons is around 70. It falls under the slowest Pokemon among all the first gen Pokemons.
