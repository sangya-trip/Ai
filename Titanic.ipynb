```python
import pandas as pd

df = pd.read_csv('titanic.csv') 
female_passengers = df[df['Sex'] == 'female']
total_females = len(female_passengers)
female_survivors = len(female_passengers[female_passengers['Survived'] == 1])
female_survival_percentage = (female_survivors / total_females) * 100
print(f"Percentage of female passengers who survived: {female_survival_percentage:.2f}%")
```
