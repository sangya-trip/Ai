```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# -------------------------------
# Step 1: Set the correct file path
# Replace this with your actual file location
# Example:
# Windows: "C:/Users/Kate/Documents/Denver Ramen Restaurant.csv"
# Mac/Linux: "/Users/Kate/Documents/Denver Ramen Restaurant.csv"
file_path = "C:/path/to/your/Denver Ramen Restaurant.csv"
# -------------------------------

# Load the dataset
file_path = "Denver Ramen Restaurants.csv"
df = pd.read_csv(file_path)

# Strip whitespace from column names (important if CSV has extra spaces)
df.columns = df.columns.str.strip()

# Convert Yelp and Google ratings to numeric, invalid entries become NaN
df['Yelp Rating'] = pd.to_numeric(df['Yelp Rating'], errors='coerce')
df['Google Rating'] = pd.to_numeric(df['Google Rating'], errors='coerce')

# Drop rows with missing values
df = df.dropna(subset=['Yelp Rating', 'Google Rating'])

# Extract relevant columns
x = df['Yelp Rating'].values.reshape(-1, 1)  # Independent variable (Yelp)
y = df['Google Rating'].values.reshape(-1, 1)  # Dependent variable (Google)

# Build a linear regression model
model = LinearRegression()
model.fit(x, y)

# Predict the Google rating for Taki Sushi (Yelp Rating = 4.3)
taki_yelp = 4.3
taki_google_pred = model.predict(np.array([[taki_yelp]]))[0][0]

# Scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label="Restaurants", alpha=0.7)

# Line of best fit
x_range = np.linspace(min(x), max(x), 100).reshape(-1, 1)
plt.plot(x_range, model.predict(x_range), color='red', label="Line of Best Fit")

# Labels and title
plt.title("Yelp Rating vs. Google Rating (Denver Ramen Restaurants)")
plt.xlabel("Yelp Rating")
plt.ylabel("Google Rating")
plt.legend()
plt.grid(True)
plt.show()

# Print the predicted Google rating
print(f"Predicted Google Rating for Taki Sushi (Yelp 4.3): {taki_google_pred:.2f}")
```

If Taki Sushi has a Yelp rating of 4.3, its predicted Google rating is 4.6. It suggests that if the yelp review is higher, the google reviews are usually higher than the yelp reviews. It is helpful tfor people who likes to compare reviews online before going into a restaurant.
